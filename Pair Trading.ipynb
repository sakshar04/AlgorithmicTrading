```python
start_year = 2018
end_year = 2020
data = {}
for i in names:
    data[i] = yf.download(i,start=date(start_year,1,1),end=date(end_year,1,1))
    data[i] = data[i].loc[~data[i].index.duplicated(keep='first')]
```

```python
close_data = pd.DataFrame()
for i in names:
    close_data[i] = data[i]["Adj Close"]
```

```python
def find_cointegrated_pairs(data):
    n = data.shape[1]
    score_matrix = np.zeros((n, n))
    pvalue_matrix = np.ones((n, n))
    keys = data.keys()
    pairs = []
    for i in range(n):
        for j in range(i+1, n):
            S1 = data[keys[i]]
            S2 = data[keys[j]]
            result = coint(S1, S2)
            score = result[0]
            pvalue = result[1]
            score_matrix[i, j] = score
            pvalue_matrix[i, j] = pvalue
            if pvalue < 0.01:
                pairs.append((keys[i], keys[j]))
    return score_matrix, pvalue_matrix, pairs
```

```python
scores, pvalues , pairs = find_cointegrated_pairs(close_data)
```

```python
#zscor_number_1 using spread
def zspread(i):
    stock1 = close_data[i[0]]
    stock2 = close_data[i[1]]
    spread = stock1 - stock2
    slow_window = 20
    fast_window = 1
    spread_mavgf = spread.rolling(fast_window).mean()
    spread_mavgs = spread.rolling(slow_window).mean()
    spread_std = spread.rolling(slow_window).std()
    zscore = (spread_mavgf - spread_mavgs)/spread_std
    return zscore
    #zscore.plot(label = 'Rolling 20 day z score')
    #plt.axhline(0,color = 'black')
    #plt.axhline(1, c ='r',ls='--')
    #plt.axhline(-1, c ='r',ls='--')
```

```python
def zratio(i):
    stock1 = close_data[i[0]]
    stock2 = close_data[i[1]]
    spread = stock1/stock2
    slow_window = 20
    fast_window = 1
    spread_mavgf = spread.rolling(fast_window).mean()
    spread_mavgs = spread.rolling(slow_window).mean()
    spread_std = spread.rolling(slow_window).std()
    zscore = (spread_mavgf - spread_mavgs)/spread_std
    return zscore
    #zscore.plot(label = 'Rolling 20 day z score')
    #plt.axhline(0,color = 'black')
    #plt.axhline(1, c ='r',ls='--')
    #plt.axhline(-1, c ='r',ls='--')
```

```python
test_data = {}
for i in pairs:
    df = pd.DataFrame()
    df[i[0]] = close_data[i[0]]
    df[i[1]] = close_data[i[1]]
    df["zscore_spread"] = zspread(i)
    df["zscore_ratio"] = zratio(i)
    test_data[i[0] + "-" + i[1]] = df
```
